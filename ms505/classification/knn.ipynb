```python
import numpy as np
```

```python
def euclidean_distance(img1, img2):
    dist = 0
    for i in range(img1.shape[0]):
        for j in range(img1.shape[1]):
            diff = img1[i, j] - img2[i, j]
            dist += diff ** 2
    return np.sqrt(dist)
```

Example of funcionality


```python
img1 = np.array([[2, 4, 5], [4, 5, 6]])
img2 = np.array([[1, 2, 3], [4, 5, 6]])
distance = euclidean_distance(img1, img2)
print("Euclidean distance:", distance)
```
