```python
def euclidean_distance(img1, img2):
    dist = 0
    for i in range(img1.shape[0]):
        for j in range(img1.shape[1]):
            diff = img1[i, j] - img2[i, j]
            dist += diff ** 2
    return np.sqrt(dist)
```

```python
def new_euclidean(array1, array2):
    dist = 0
    for i in range(array1.shape[0]):
        diff = array1[i] - array2[i]
        dist += diff ** 2
    return np.sqrt(dist)
```

```python
array_test1 = np.array([1, 2, 3])
array_test2 = np.array([1, 3, 4])

print(new_euclidean(array_test1, array_test2))
```

```python
results = []
for k in range(len(keys_descritores_teste)):
    keys_descritores_teste = list(descritores_teste.keys())
    keys_descritores_treino = list(descritores_treino.keys())

    array_teste = descritores_teste[keys_descritores_teste[k]]
    minimum = [np.inf, ""]
    for i in keys_descritores_treino:
        dist = new_euclidean(array_teste, descritores_treino[i])
        if dist <= minimum[0]:
            minimum = [dist, i] 
    results.append((keys_descritores_teste[k], minimum[0], minimum[1]))

for r in results:
    print(f"Imagem teste: {r[0]} | Distncia: {r[1]} | Imagem treino associada: {r[2]}")
```

```python
def verify_class(results):
    correct_test = []
    incorrect_test = []
    correct_val = []
    incorrect_val = []
    for r in results:
        if r[0][3] == r[2][3] and r[0][2] == r[2][2]:
            # Mensagem formal para classificaes corretas
            print(f"Classificao correta  Imagem de teste: {r[0]}; Imagem de treino correspondente: {r[2]}")
            correct_test.append(r[0])
            correct_val.append(r[2])
        else:
            # Mensagem formal para classificaes incorretas
            print(f"Classificao incorreta  Imagem de teste: {r[0]}; Imagem de treino associada: {r[2]}")
            incorrect_test.append(r[0])
            incorrect_val.append(r[2])
    return correct_test, incorrect_test, correct_val, incorrect_val
```

```python
correct_test, incorrect_test, correct_val, incorrect_val = verify_class(results)
```

```python
print(f"Imagens corretamente classificadas: {correct_test}")
print(f"Imagens incorretamente classificadas: {incorrect_test}")
print(f"Nmero total de classificaes corretas (teste): {len(correct_test)}")
print(f"Nmero total de classificaes incorretas (teste): {len(incorrect_test)}")
```

```python
# Impresso acadmica das imagens incorretas e suas correspondentes associadas de treino
if len(incorrect_test) > 0:
    print("Lista de imagens incorretamente classificadas e respectivas imagens de treino associadas:")
    for teste, associado in zip(incorrect_test, incorrect_val):
        print(f"  Imagem de teste: {teste}  -->  Imagem de treino associada: {associado}")
else:
    print("Nenhuma classificao incorreta detectada no conjunto de teste.")
```
